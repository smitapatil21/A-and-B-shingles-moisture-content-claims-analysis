Check the normality using Empirical Rule

```python
def empirical (x):
  sd=x.std()
  mean=x.mean()
  print('68 % of data should lie between {} and {}'.format(mean-sd,mean+sd))
  print('{} % of data lies between LL and UL for 68%'.format(pd.Series((x > mean-sd) & (x< mean+sd)).value_counts(normalize=True).values[0]*100),'\n')
  print('95 % of data should lie between {} and {}'.format(mean-(2*sd),mean+(2*sd)))
  print('{} % of data lies between LL and UL for 95%'.format(pd.Series((x > mean-(2*sd)) & (x< mean+(2*sd))).value_counts(normalize=True).values[0]*100),'\n')
  print('99 % of data should lie between {} and {}'.format(mean-(3*sd),mean+(3*sd)))
  print('{} % of data lies between LL and UL for 99%'.format(pd.Series((x > mean-(3*sd)) & (x< mean+(3*sd))).value_counts(normalize=True).values[0]*100),'\n')
```

Its tedious to caluculate mean and standard devaition each time so we define a function empirical and input x .Calculating mean and standard deviation of x and stored in mean and sd .
The 68% of data lies between mean -1 std and + 1 std deviation and checked how many data points actually lies within this range .
Data point which lies  within lower limit being -1 std and +1 std as upper limit .
Series created value x greater than _1 sd and lesser than +1 std ,passing value counts .Since this is logical operator it gives bullian values of True and false .We are counting true and false values .
Normalise true will give the proportion of true values in series and taking the 0th value multiplying by 100.Means we are taking only true values multiplied by 100.

Empirical Rule states that for a normal distribution ,nearly all of the data point will fall within the standard deviation of mean .
The empirical rule can be broken down into three parts .

Problem 

An important quality characteristic used by the manufacturers of ABC asphalt shingles is the amount of moisture the shingles contain when they are packaged. Customers may feel that they have purchased a product lacking in quality if they find moisture and wet shingles inside the packaging.   In some cases, excessive moisture can cause the granules attached to the shingles for texture and colouring purposes to fall off the shingles resulting in appearance problems. To monitor the amount of moisture present, the company conducts moisture tests. A shingle is weighed and then dried. The shingle is then reweighed, and based on the amount of moisture taken out of the product, the pounds of moisture per 100 square feet is calculated. The company claims that the mean moisture content cannot be greater than 0.35 pound per 100 square feet.
The file (A & B shingles.csv) includes 36 measurements (in pounds per 100 square feet) for A shingles and 31 for B shingles.

For the A shingles, the null and alternative hypothesis to test whether the population mean moisture content is less than 0.35 pound per 100 square feet  is given:

LaTeX: H_0<=0.35

LaTeX: H_A>0.35

For the B shingles, the null and alternative hypothesis to test whether the population mean moisture content is less than 0.35 pound per 100 square feet is given:

LaTeX: H_0<=0.35

LaTeX: H_A>0.35

For A shingles-Hypothesis:

Test the Hypothesis 

Step 1: Define null and alternative hypotheses
In testing the average minutes spent on the internet.

Null hypothesis :Company claims or status quo . Only under strong evidence, the null hypothesis is rejected. Null Hypothesis states that the mean moisture content ,mu is less than equals to 0.35 pound per 100 square feet.
Alternative hypotheis :mu is greater than 0.35 pounds per 100 square feet . Point noted here is "=" sign must be included only in null hypothesis and never in alternative hypothesis.

H0:  <= 0.35
HA:  > 0.35

The test for population mean is done using 1 sample t test . Since the alternative hypothesis is greater ,the rejection region will be on right hand side . If the p-value is less than 0.05 ,the null hypothesis is rejected .

Step 2: Decide the significance level
Here we select  = 0.05.

```python
print("The sample size for this problem is",len(df.A))
```

Step 3: Identify the test statistic
We do not know the population standard deviation and n = 36. So we use the t distribution and the tSTAT test statistic.

Step 4: Calculate the p - value and test statistic
scipy.stats.ttest_1samp calculates the t test for the mean of one sample given the sample observations and the expected value in the null hypothesis. This function returns t statistic and the two-tailed p value.

```python
# one sample t-test
# null hypothesis: expected value = 0.35
# By default t-test is two-sided t-test in python
t_statistic, p_value = stats.ttest_1samp(df.A.dropna(), 0.35)
print('One sample t test \nt statistic: {0} p value: {1} '.format(t_statistic, p_value))
```

Let us calculate at the value of the test statistic manually

Step 5 Decide to reject or accept null hypothesis

Python gives us two sided test for null hypothesis .But here this is one -sided test  as per the formulated hypothesis .
How to force python to calculate alternative hypotheis which is greater side or one sided.
We are looking here for rejection region on right hand side(one sided) .
Area under the curve of t-distribution =1,we are subtracting 1 from p-value/2 to get the right sided value.
Given p_value will be two sided p_value.Hence dividing p -value by 2 and substracting it from 1 will give the p -value for right side of t distribution.

Lets check if p-value is less than alpha or not.Here the condition is if the calculated p-value is greater than alpha .
Since the alternate in our case lies on right side 

alpha_value = 0.05 # Level of significance



```python
alpha_value = 0.05
```

```python
p_value_greater=1-(p_value/2)
p_value_greater
```

```python
# one sample t-test
# null hypothesis: expected value = 0.35
# By default t-test is two-sided t-test in python
t_statistic, p_value = stats.ttest_1samp(df.A, 0.35)
print('One sample t test \nt statistic: {0} p value: {1} '.format(t_statistic, p_value))

p_value_greater=1-(p_value/2)

if p_value_greater > 0.05:
  print('The p_value which is greater than the level of significance 0.05,hence we fail to reject the null hypothesis',format(p_value_greater))

else:
  
    print('The p_value which is less than the level of significance 0.05,hence we reject the null hypothesis',format(p_value_greater)) 

  
print ("Our one-sample t-test p-value=",p_value_greater)
```

Conclusion :
The p-value is 0.925 which is greater than the level of significance 0.05 ,hence we fail to reject the null hypothesis.
Its concluded that for A shingles the null hypothesis that the mean moisture content is less than 0.35 cannot be rejected.

3.1 Do you think that the population means for shingles A and B are equal? Form the hypothesis and conduct the test of the hypothesis. What assumption do you need to check before the test for equality of means is performed?

Null Hypothesis states that the population mean of A and  shingles are equal 


H0:A=B

Alternative hypothesis states that the population mean of A and B shingles are not equal.
Ha:A!=B


Assumption for t-test:
1.The first assumption made concerns the scale of measurement-The assumption for t-test is that the scale of measuremnt applied to the data follows a continous or ordinal scale.
2.The second assumption is that the data is randomnly selected from total poplulation-Random sample.
3.The third assumption is distribution of moisture content in both population follow the normal distribution.
4.The fourth assumption is large sample size is used.
5.The final assumption is homogeneity of variance -samples with equal variance .


```python

# one sample t-test
# null hypothesis: expected value = 0.35
# By default t-test is two-sided t-test in python
t_statistic, p_value = stats.ttest_ind(df['A'],df['B'],axis=0,equal_var=True,nan_policy='omit')
print('Two sample t test \nt statistic: {0} p value: {1} '.format(t_statistic, p_value))
```

```python
alpha_value=0.05
# level of significance
print('Level of significance: %.2f'%alpha_value,'\n')
```

```python


if p_value < 0.05:
  print('We have evidence to reject the null hypothesis since p -value is < level of significance')

else:
  
    print('We have  no evidence to reject the null hypothesis since p-value is greater than level of significance') 
```

Thus we conclude that the two population mean are indeed equal.
